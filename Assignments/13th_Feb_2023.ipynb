### Q1. Explain why we have to use the Exception class while creating a custom Exception.

2. it makes it easy to customize the error message and provide additional information about the error.
3. Code will be more easier to understand for other developers who are familiar with Python's error handling mechanisms.

```python
# code to demonstrate the use of Exception class. 

try:
    a=10
    a/0
except Exception as e:   # by using Exception class it will identify the error.
    print("This is error while running the code." , e)
```

### Q3. What error are defined in ArithmeticError class ? explain with two example..
    The ArithmeticError Exception is the base class for all errors associated with arithmetic operation
1. ZeroDivisionError : raised when we try to divide any number with Zero.
2. OverflowError : raised when arithmetic operation is too large to be represented.
3. FloatingPointError

```python
# code to demonstrate the ZeroDivisionError

try:
    1/0   # dividing number with zero
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
# code to demonstrate the OverflowError

num = 5.0

try:
    for i in range(1, 1000):
        num = num ** i    # it will end up in very large value of number....
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

### Q4. Why lookupError class is used? Explain with an example KeyError and IndexError.
    LookupError is raised when a key or index used on a mapping or sequence is invalid.
1. IndexError: raised when index is out of range 
2. KeyError: raised when value for that key is not present in the dict.

```python
# code to demonstrate the IndexError

x = [1, 2, 3, 4]  # list with four element

try:
    print(x[10])  # trying to get 10th index which is not present in the list
    
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

```python
# code to demonstrate the KeyError.

course_info = {'course_name': 'Python',
                'Duration_hours': 28,
                'language': 'English'}

try:
    course_info["key_2"]   # this key does not exist in the above dict.
    
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

### Q6. List down some best practises for exception handling in python.
    an ImportError is raised in following conditions.

1. Import statement has troubles trying to load a particuler module
2. If a module does not exist.
3. If from X import Y is used and Y cannot be found inside the module X

#### Ans. Here is the list of best practises for exception handling in python

####  1. Use always specific exception.
####  2. print always a valid and proper message.
####  3. always try to do logging. Dont use Print statement.
####  4. always avoid to write a multiple exception handling.
####  5. always try to prepare/maintain proper documentation.
####  6. cleanup all the resources after using the file..


```python
# 1. use always specific exception..

try:
    a=20
    a/0
except ZeroDivisionError as e:
    print(e)
```

```python
# 2. print always a valid massage

try:
    a=20
    a/0
except ZeroDivisionError as e:
    print("this is zero division error.........",e)
```

```python
# 3. always try to log....dont use print

import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)

try:
    10/0
except ZeroDivisionError as e:
    logging.error("this is my zero division error i am handling {} ".format(e))
```

```python
# 4. always avoid to write a multiple exception handling.......
try:
    10/0
except FileNotFoundError as e:
    logging.error("this is my file not found error {} ".format(e))
except ZeroDivisionError as e:
    logging.error("this is my zero division error i am handling {} ".format(e))
```

```python
# 6. cleanup all the resources....

try:
    with open("test987.txt","w") as f:
        f.write("this is my msg to file.....")
except FileNotFoundError as e:
    logging.error("this is my file not found error {} ".format(e))
finally:
    f.close()     # need to clean the resources...
    
```
